```
df_new = df_ind_new[['date','new_cases']]
```

```
df_new = df_new.rename(columns={'date':'ds','new_cases':'y'})
df_new
```

```
import fbprophet
model = fbprophet.Prophet(changepoint_prior_scale=0.7)
model.fit(df_new)
```

```
forecast = model.make_future_dataframe(periods = 10,freq='M')
forecast = model.predict(forecast)
```
